```python
#ratings = pd.read_csv("/content/drive/MyDrive/BX-Book-Ratings.csv",sep=";",error_bad_lines=False,encoding="latin-1")
ratings = pd.read_csv("BX-Book-Ratings.csv",sep=";",error_bad_lines=False,encoding="latin-1")
```

```python
ratings.head()
```

```python
ratings = ratings.rename(columns = {"User-ID":"user_id","Book-Rating":"rating"})
```

```python
ratings.head()
```

```python
books.shape
```

```python
users.shape
```

```python
ratings.shape
```

```python
x = ratings['user_id'].value_counts()>200
```

```python
y = x[x].index
```

```python
y
```

```python
ratings = ratings[ratings['user_id'].isin(y)]
```

```python
ratings.shape
```

```python
ratings.head()
```

```python
ratings_with_books = ratings.merge(books,on="ISBN")
```

```python
ratings_with_books
```

```python
number_of_ratings = ratings_with_books.groupby("title")['rating'].count().reset_index().rename(columns={"rating":"No. of ratings"})
```

```python
number_of_ratings.head()
```

```python
final_rating = ratings_with_books.merge(number_of_ratings, on="title")
```

```python
final_rating.head()
```

```python
final_rating.shape
```

```python
final_rating = final_rating[final_rating["No. of ratings"]>=50]
```

```python
final_rating.shape
```

```python
final_rating.drop_duplicates(["user_id","title"], inplace = True)
```

```python
final_rating.shape
```

```python
book_pivot = final_rating.pivot_table(columns="user_id", index= "title", values = "rating")
```

```python
book_pivot
```

```python
book_pivot.shape
```

```python
book_pivot.fillna(0, inplace=True)
```

```python
book_pivot
```

```python
from scipy.sparse import csr_matrix
book_sparse = csr_matrix(book_pivot)
```

```python
type(book_sparse)
```

```python
from sklearn.neighbors import NearestNeighbors
```

```python
model = NearestNeighbors(algorithm = "brute")
```

```python
model.fit(book_sparse)
```

```python
distances, suggestion = model.kneighbors(book_pivot.iloc[237,:].values.reshape(1,-1),n_neighbors=6)
```

```python
distances
```

```python
suggestion
```

```python
for i in range(len(suggestion)):
  print(book_pivot.index[suggestion[i]])
```

```python
def recommend_book_similar_to(book_name):
  book_id = np.where(book_pivot.index==book_name)[0][0]
  distances, suggestion = model.kneighbors(book_pivot.iloc[book_id,:].values.reshape(1,-1),n_neighbors=6)

  for i in range(len(suggestion)):
    print(book_pivot.index[suggestion[i]])
```

```python
recommend_book_similar_to("Animal Farm")
```
